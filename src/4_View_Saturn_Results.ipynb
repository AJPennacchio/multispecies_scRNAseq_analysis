```python
import scanpy as sc
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def view_integration(adata):
    sc.pp.pca(adata)
    sc.pp.neighbors(adata)
    sc.tl.umap(adata)
    sc.pl.umap(adata, color="species", title="UMAP by Species", size=10)
    sc.pl.umap(adata, color="labels2", title="UMAP by Integration Column", size=10)
    sc.pl.umap(adata, color="ref_labels", title="UMAP by Reference", size=10)
```

```python
#UPDATE PATH to SATURN output, should look something like "test256_data_SAMPLE_NAMES_org_saturn_seed_0.h5ad"
adata = sc.read("/home/apennacchio/pcsct/saturn/out/14/test256_data_sbicolor_root10_integrated_zmays23_org_saturn_seed_0.h5ad")
view_integration(adata)
```
